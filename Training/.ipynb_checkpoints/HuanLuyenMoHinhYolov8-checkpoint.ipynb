```python
import os
import random
import shutil

# Đường dẫn thư mục chứa dữ liệu ban đầu
input_folder = r"G:/Detected-Number/labelImg/data/GreenParking"

# Đường dẫn thư mục output
output_folder = r"F:/jupyter/PhatHienBienSoXe/data"
images_train_folder = os.path.join(output_folder, "images/train")
images_val_folder = os.path.join(output_folder, "images/val")
labels_train_folder = os.path.join(output_folder, "labels/train")
labels_val_folder = os.path.join(output_folder, "labels/val")

# Tạo các thư mục nếu chưa tồn tại
os.makedirs(images_train_folder, exist_ok=True)
os.makedirs(images_val_folder, exist_ok=True)
os.makedirs(labels_train_folder, exist_ok=True)
os.makedirs(labels_val_folder, exist_ok=True)

# Lấy danh sách tất cả file .jpg
image_files = [f for f in os.listdir(input_folder) if f.endswith('.jpg')]

# Shuffle dữ liệu để đảm bảo tính ngẫu nhiên
random.seed(1337)
random.shuffle(image_files)

# Chia dữ liệu thành 80% train, 20% val
split_index = int(0.8 * len(image_files))
train_files = image_files[:split_index]
val_files = image_files[split_index:]

# Hàm di chuyển file và nhãn tương ứng
def move_files(file_list, source_folder, dest_image_folder, dest_label_folder):
    for file_name in file_list:
        # Di chuyển file ảnh
        src_image = os.path.join(source_folder, file_name)
        dest_image = os.path.join(dest_image_folder, file_name)
        shutil.copy(src_image, dest_image)

        # Di chuyển file nhãn tương ứng
        label_name = file_name.replace('.jpg', '.txt')
        src_label = os.path.join(source_folder, label_name)
        dest_label = os.path.join(dest_label_folder, label_name)
        if os.path.exists(src_label):
            shutil.copy(src_label, dest_label)

# Di chuyển file vào train và val
move_files(train_files, input_folder, images_train_folder, labels_train_folder)
move_files(val_files, input_folder, images_val_folder, labels_val_folder)

print("Dữ liệu đã được chia thành công!")
```

```python
import cv2
import os
import matplotlib.pyplot as plt
from ultralytics import YOLO

# Load mô hình YOLOv8 đã huấn luyện
model = YOLO("F:/jupyter/PhatHienBienSoXe/model/best-5.pt")

def preprocess_image(image_path):
    # Đọc ảnh
    image = cv2.imread(image_path)
    if image is None:
        print("Không thể đọc ảnh. Vui lòng kiểm tra lại đường dẫn ảnh.")
        return None, None

    # Resize ảnh về chiều rộng 500px và giữ tỷ lệ chiều cao
    width = 500
    height = int((width / image.shape[1]) * image.shape[0])
    image = cv2.resize(image, (width, height))

    # Chuyển ảnh sang ảnh xám
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    # Áp dụng Gaussian Blur để lọc nhiễu
    gray = cv2.GaussianBlur(gray, (5, 5), 0)
    
    # Cân bằng histogram để làm nổi bật chi tiết
    gray = cv2.equalizeHist(gray)
    
    # Áp dụng Bilateral Filter để giữ các cạnh sắc nét
    gray = cv2.bilateralFilter(gray, 11, 17, 17)
    
    return gray, image

def detect_license_plate(image):
    # Dự đoán với mô hình YOLOv8
    results = model(image)
    
    # Lấy các bounding boxes của biển số xe
    boxes = results[0].boxes.xyxy.cpu().numpy() if results else []
    if len(boxes) == 0:
        print("Không phát hiện biển số xe trong ảnh.")
        return None

    return boxes

def crop_license_plate(gray_image, image, boxes, save_path):
    os.makedirs(save_path, exist_ok=True)
    cropped_images = []

    for i, box in enumerate(boxes):
        x1, y1, x2, y2 = map(int, box)
        
        # Cắt ảnh theo bounding box (dùng ảnh xám đã xử lý)
        cropped_image = gray_image[y1:y2, x1:x2]
        
        if cropped_image is None or cropped_image.size == 0:
            print(f"Không thể cắt ảnh tại bounding box {i+1}.")
            continue

        # Lưu ảnh đã cắt vào thư mục
        output_file = os.path.join(save_path, f"cropped_image_{i+1}.jpg")
        if cv2.imwrite(output_file, cropped_image):
            print(f"Ảnh đã được lưu vào {output_file}")
        else:
            print(f"Lỗi khi lưu ảnh tại bounding box {i+1}.")
        
        # Lưu ảnh đã cắt vào danh sách
        cropped_images.append(cropped_image)
    
    return cropped_images

# Đọc ảnh và tiền xử lý
image_path = "F:/jupyter/PhatHienBienSoXe/samples/218.jpg"
gray, image = preprocess_image(image_path)

# Dự đoán biển số xe
boxes = detect_license_plate(image)

# Nếu phát hiện biển số xe, tiến hành cắt ảnh
if boxes is not None:
    save_path = "F:/jupyter/PhatHienBienSoXe/results/"
    cropped_images = crop_license_plate(gray, image, boxes, save_path)

    # Hiển thị ảnh đầu vào và các ảnh xử lý
    fig, axes = plt.subplots(2, 3, figsize=(12, 8))  # Tạo lưới 2x3

    # Danh sách ảnh để hiển thị
    images_to_show = [
        image,   # Ảnh gốc
        gray,    # Ảnh xám
        cv2.GaussianBlur(gray, (5, 5), 0),  # Ảnh lọc nhiễu
        cv2.equalizeHist(gray),  # Ảnh cân bằng histogram
        cv2.bilateralFilter(gray, 11, 17, 17),  # Ảnh sau khi áp dụng bilateral filter
        cropped_images[0] if len(cropped_images) > 0 else gray  # Ảnh cắt (nếu có)
    ]

    # Đảm bảo rằng số ảnh là 6, nếu ít hơn, chúng ta sẽ để trống các ô còn lại
    for i, ax in enumerate(axes.flat):
        if i < len(images_to_show):
            ax.imshow(images_to_show[i], cmap='gray' if len(images_to_show[i].shape) == 2 else None)
            ax.set_title(f"Image {i+1}")
        ax.axis('off')  # Tắt hiển thị trục

    plt.tight_layout()  # Điều chỉnh layout để tránh trùng lắp
    plt.show()
else:
    print("Không phát hiện biển số xe nào trong ảnh.")
```

```python
#Nhan dien video 
from ultralytics import YOLO
import cv2
from pathlib import Path

def process_video(video_path, output_path, model_path="F:/jupyter/PhatHienBienSoXe/model/best.pt"):
    # Load the YOLO model
    model = YOLO(model_path)
    
    # Open the video file
    video = cv2.VideoCapture(video_path)
    
    # Get video properties
    width = int(video.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(video.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps = int(video.get(cv2.CAP_PROP_FPS))
    
    # Create video writer
    output = cv2.VideoWriter(
        output_path,
        cv2.VideoWriter_fourcc(*'mp4v'),
        fps,
        (width, height)
    )
    
    # Process the video frame by frame
    while video.isOpened():
        success, frame = video.read()
        
        if not success:
            break
            
        # Run YOLO detection on the frame
        results = model(frame, stream=True)
        
        # Process detection results
        for result in results:
            # Plot the results on the frame
            annotated_frame = result.plot()
            
            # Write the frame with detections to the output video
            output.write(annotated_frame)
    
    # Release resources
    video.release()
    output.release()
    cv2.destroyAllWindows()
    
    print(f"Video processing completed. Output saved to: {output_path}")

# Example usage
if __name__ == "__main__":
    input_video = "F:/jupyter/PhatHienBienSoXe/samples/0111.mp4"  # Đường dẫn đến video đầu vào
    output_video = "F:/jupyter/PhatHienBienSoXe/results/output_video.mp4"  # Đường dẫn lưu video kết quả
    process_video(input_video, output_video)
```

```python
import os

# Đường dẫn tới thư mục chứa các file
folder_path = 'G:/Detected-Number/labelImg/data/GreenParking'

# Lấy danh sách các file trong thư mục
files = os.listdir(folder_path)

# Lọc ra các file có phần mở rộng .jpg và .txt
jpg_files = [f for f in files if f.endswith('.jpg')]
txt_files = [f for f in files if f.endswith('.txt')]

# Sắp xếp danh sách file theo tên (hoặc theo thứ tự mong muốn)
jpg_files.sort()
txt_files.sort()

# Đảm bảo rằng số lượng file jpg và txt là giống nhau
min_length = min(len(jpg_files), len(txt_files))

# Đổi tên file theo thứ tự
for i in range(min_length):
    # Tạo tên mới cho các file
    new_name = f"{1020+i}"
    
    # Đổi tên các file .jpg
    old_jpg_name = os.path.join(folder_path, jpg_files[i])
    new_jpg_name = os.path.join(folder_path, new_name + '.jpg')
    os.rename(old_jpg_name, new_jpg_name)
    
    # Đổi tên các file .txt
    old_txt_name = os.path.join(folder_path, txt_files[i])
    new_txt_name = os.path.join(folder_path, new_name + '.txt')
    os.rename(old_txt_name, new_txt_name)

print("Đổi tên file thành công!")
```
